# Importing libraries


```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer #used to get the count of each word
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
```

```python
import nltk
nltk.download('stopwords')
```

```python
import re #string matching using regular expression 
import nltk #NLP library for added functionality 
from nltk.util import pr 
stemmer =nltk.SnowballStemmer("english")
from nltk.corpus import stopwords 
import string
stopword=set(stopwords.words("english"))

```

```python
df=pd.read_csv("/twitter_data.csv")
```

```python
print(df.head())
```

```python
df["labels"]=df['class'].map({0:"profanity detected",1:"profanity detected",2:"no profanity detected",3:"no profanity detected"})
```

```python
print(df.head())
```

```python
df=df[['tweet','labels']]
df.head()
```

# cleaning the dataset 

```python
def clean(text):
  text=str(text).lower()
  text=re.sub('\[.*?\]','',text)
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('<.*?>+','',text)
  text=re.sub('[%s]'% re.escape(string.punctuation),'',text)
  text=re.sub('\n','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  text=[stemmer.stem(word) for word in text.split(' ')]
  text=" ".join(text)
  return text
df["tweet"]=df["tweet"].apply(clean)  
print(df.head())


```

```python
 df['tweet'] = df['tweet']. fillna(0)
 df['tweet']= df['tweet']
```

```python
df['labels']=df['labels'].fillna(0)
df['labels']=df['labels']
```

```python
x=np.array(df["tweet"])
y=np.array(df["labels"])
cv= CountVectorizer()
x=cv.fit_transform(x)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)



```

```python
for i in y_train:
  print(i)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

# Training data using decision tree classifier 


```python
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
for i in y_train:
  print(i)
```

```python
testtext="you are not someone I would like to talk to "
df=cv.transform([testtext]).toarray()
df=clf.predict(df)
if (df==1):
  print("profanity detected")
else:
  print("profanity not detected")

```
